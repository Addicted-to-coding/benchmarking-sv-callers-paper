```python
#concatenating all tools and strains at threshold 0, 10, 100, 1000, and 10000
import os
import allel
import pandas as pd

#missing popdel files, not included in tools list
strain=['A_J','AKR_J','BALB_CJ','CBA_J','C3H_HeJ','DBA_2J','LP_J']
tools=['BioGraph*',
'breakdancer',
'clever',
'delly',
'gasv',
'gridss',
'indelminer',
'lumpexpress',
'mistrvar',
'pindel',
'platypus',
'rdxplorer',
'smoove',
'sniffles']

#0 threshold
th=0
df_current = pd.DataFrame(columns=['strain','length','flag','position','threshold'])
df_0 = pd.DataFrame(columns=['tool','strain','length','flag','position','threshold'])
for s in strain:
    for t in tools:
        file='../raw_data/mouse/custom_vcf_fulldata/'+str(th)+'t/nf_'+str(th)+'t.'+t+'.'+s+'.chr19.100p_sorted.modified.vcf'
        
        callset = allel.read_vcf(file,fields='*')
        if callset!=None:
            df_current = pd.DataFrame({'tool': t, 'strain': s, 'length': callset['variants/SVLEN'],'flag': callset['variants/FLAG'],'position': callset['variants/POS'],'threshold': th})
            df_0 = pd.concat([df_current, df_0],ignore_index=True)   
        
df_0.head()
```
